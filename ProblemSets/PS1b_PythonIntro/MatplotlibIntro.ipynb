```python
# import packages
import numpy as np
from matplotlib import pyplot as plt
```

### Problem 1

```python
# define functions
def varmean(n):
    stdnorm = np.random.normal(size=(n,n))
    rowmean = np.mean(stdnorm, axis=1)
    return np.var(rowmean)

def result(vecn):
    res = np.zeros(len(vecn))
    for n in vecn:
        res[int(n/100)-1] = varmean(n)
    return res

# apply functions
x = np.arange(100,1100,100)
y = result(x)

# plot results
plt.plot(x,y)
plt.show()
```

### Problem 2

```python
# define domain
x = np.linspace(-2*np.pi, 2*np.pi, 200)

# plot
plt.plot(x, np.sin(x), label=r"$sin(x)$")
plt.plot(x, np.cos(x), label=r"$cos(x)$")
plt.plot(x, np.arctan(x), label=r"$arctan(x)$")

plt.legend(loc="upper left")
plt.show()
```

### Problem 3

```python
# define domain and function
x1 = np.linspace(-2,1,50, endpoint=False)
x2 = np.linspace(1,6,50)
x2 = x2[1:]

f1 = 1 / (x1 - 1)
f2 = 1 / (x2 - 1)

# plot
plt.plot(x1, f1, 'm--', linewidth=4)
plt.plot(x2, f2, 'm--', linewidth=4)
plt.xlim(-2,6)
plt.ylim(-6,6)

plt.show()
```

### Problem 4

```python
# domain
x = np.linspace(0,2*np.pi,100)

# plot sin(x)
plt.subplot(221)
plt.plot(x, np.sin(x), 'g-')
plt.axis([0, 2*np.pi, -2, 2])
plt.title(r"$sin(x)$")

# plot sin(2x)
plt.subplot(222)
plt.plot(x, np.sin(2*x), 'r--')
plt.axis([0, 2*np.pi, -2, 2])
plt.title(r"$sin(2x)$")

# plot 2sin(x)
plt.subplot(223)
plt.plot(x, 2*np.sin(x), 'b--')
plt.axis([0, 2*np.pi, -2, 2])
plt.title(r"$2sin(x)$")

# plot 2sin(2x)
plt.subplot(224)
plt.plot(x, 2*np.sin(2*x), 'm:')
plt.axis([0, 2*np.pi, -2, 2])
plt.title(r"$2sin(2x)$")

plt.suptitle("Plots for problem 4")
```
