```python
# import packages
import numpy as np
from matplotlib import pyplot as plt
import scipy.special
```

### Problem 2

```python
#Â define axis x
x = np.linspace(0,1,200)

plt.figure(figsize = (12,8))
for n in range(4):
    for v in range(n+1):
        b = scipy.special.binom(n,v) * x**v * (1-x)**(n-v)
        plt.subplot(4,4,(4*n+1)+v)
        plt.subplots_adjust(hspace = .5)
        plt.subplots_adjust(wspace = .5)
        plt.plot(x, b)
        plt.title(r"$b_{}$".format(n)+r"$_,$"+r"$_{}(x)$".format(v))
        plt.axis([-0.1, 1.1, -0.1, 1.1])

plt.show()
```

### Problem 3

```python
from numpy.polynomial.polynomial import polyfit

# load data
info = np.load("MLB.npy")

# weight - height
b, m = polyfit(info[:,1], info[:,0], 1)
plt.scatter(info[:,1], info[:,0], alpha=.5, edgecolor='none')
plt.plot(info[:,1], b + m * info[:,1], 'r-')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.axis('equal')
plt.title("Correlation weight - height")
plt.show()

# height - age
b, m = polyfit(info[:,0], info[:,2], 1)
plt.scatter(info[:,0], info[:,2], alpha=.5, edgecolor='none')
plt.plot(info[:,0], b + m * info[:,0], 'r-')
plt.xlabel("Height")
plt.ylabel("Age")
plt.axis('equal')
plt.title("Correlation height - age")
plt.show()

# weight - age
b, m = polyfit(info[:,1], info[:,2], 1)
plt.scatter(info[:,1], info[:,2], alpha=.5, edgecolor='none')
plt.plot(info[:,1], b + m * info[:,1], 'r-')
plt.xlabel("Weight")
plt.ylabel("Age")
plt.axis('equal')
plt.title("Correlation weight - age")
plt.show()
```

### Problem 5

```python
#Â define domain and function
x = np.linspace(-2,2,200)
X, Y = np.meshgrid(x, x)
Z = (1-X)**2 + 100*(Y-X**2)**2

fig, ax = plt.subplots(1,2)
# plot heat map
cax = ax[0].pcolormesh(X, Y, Z, cmap='inferno')
ax[0].plot(1,1,'wo')
fig.colorbar(cax, ax=ax[0])

# plot contour
cax = ax[1].contour(X, Y, Z, [100, 300, 500, 1000, 3000, 4000], cmap='viridis')
ax[1].plot(1,1,'ko')
fig.colorbar(cax, ax=ax[1])

fig.tight_layout()
fig.suptitle("Rosenbrock function")
fig.subplots_adjust(top=0.88)
plt.show()
```

### Problem 6

```python
# load data
data = np.load("countries.npy")
countries = ["Austria", "Bolivia", "Brazil", "China",
            "Finland", "Germany", "Hungary", "India",
            "Japan", "North Korea", "Montenegro", "Norway",
            "Peru", "South Korea", "Sri Lanka", "Switzerland",
            "Turkey", "United Kingdom", "United States", "Vietnam"]

# scatter plot population-GDP
plt.scatter(np.log(data[:,0]), np.log(data[:,1]))
plt.xlabel("log(population)")
plt.ylabel("log(GDP)")
plt.title("Correlation population - GDP")
plt.show()

# bar chart GDP
fig, ax = plt.subplots()
positions = np.argsort(data[:,1])
countries = np.array(countries)
countries = countries[positions]
ax.barh(countries, np.sort(data[:,1]), align="center")
ax.set_yticks(countries)
plt.title("GDP")
plt.show()

# histogram average height for male and female
plt.subplot(121)
plt.hist(data[:,2], bins=10)
plt.title("Average male height (cm)")
plt.subplot(122)
plt.hist(data[:,3], bins=10)
plt.title("Average female height (cm)")
plt.show()
```

From the first graph we can see that population and GDP are positively correlated. Bigger countries in term of population have on average a higher GDP.

The bar chart shows how the GDP is distributed among countries. The three largest economies are the United States, China, and Japan.

The two histograms exhibit the distribution of the average male and female height worldwide. The obvious difference is that on average men are taller than women. Funny fact: both the instograms present the lowest average height for Bolivian people!
