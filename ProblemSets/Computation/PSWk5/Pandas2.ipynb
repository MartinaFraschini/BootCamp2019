```python
# import packages
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
```

### Problem 1 - Road dataset

```python
# load data
df = data("birthdeathrates")
data("birthdeathrates", show_doc=True)
```

```python
df.head()
```

```python
df.plot(linewidth=1, title="Database overview")
plt.show()
```

```python
pd.plotting.scatter_matrix(df[["birth", "death"]])
plt.show()
```

```python
df.plot(kind="scatter",x="birth",y="death",title="idk")
plt.show()
```

```python
df[["birth","death"]].hist(grid=False,bins=20)
plt.show()
```

```python
df.plot(kind="box",vert=False)
plt.show()
```
